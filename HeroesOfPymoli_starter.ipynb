```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players
* Percentage and Count of Female Players
* Percentage and Count of Other / Non-Disclosed

```python
#Percentage and count by gender
male_players_df=purchase_df.loc[purchase_df['Gender'] == 'Male']
male_players_count = male_players_df['SN'].nunique()
male_players_percent=round((male_players_count/total_no_of_players)*100,2)
female_players_df=purchase_df.loc[purchase_df['Gender'] == 'Female']
female_players_count = female_players_df['SN'].nunique()
female_players_percent=round((female_players_count/total_no_of_players)*100,2)
other_players_df=purchase_df.loc[purchase_df['Gender'] == 'Other / Non-Disclosed']
other_players_count=other_players_df['SN'].nunique()
other_players_percent=round((other_players_count/total_no_of_players)*100,2)
```

```python
#Gender demographics summary
gender_demographics=pd.DataFrame({'Total Count':[male_players_count,female_players_count,other_players_count],
                                  'Percentage of Players':[male_players_percent,female_players_percent,other_players_percent]},
                                   index=['Male','Female','Other / Non-Disclosed'])
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchase Count
male_purchase_count = purchase_df.loc[purchase_df['Gender']=='Male']['Purchase ID'].count()
female_purchase_count = purchase_df.loc[purchase_df['Gender']=='Female']['Purchase ID'].count()
other_purchase_count = purchase_df.loc[purchase_df['Gender']=='Other / Non-Disclosed']['Purchase ID'].count()
```

```python
#Average Purchase Price
avg_price_by_male='${:,.2f}'.format(purchase_df.loc[purchase_df['Gender']=='Male']['Price'].mean())
avg_price_by_female='${:,.2f}'.format(purchase_df.loc[purchase_df['Gender']=='Female']['Price'].mean())
avg_price_by_other='${:,.2f}'.format(purchase_df.loc[purchase_df['Gender']=='Other / Non-Disclosed']['Price'].mean())
```

```python
#Total Purchase Value
total_price_by_male='${:,.2f}'.format(purchase_df.loc[purchase_df['Gender']=='Male']['Price'].sum())
total_price_by_female='${:,.2f}'.format(purchase_df.loc[purchase_df['Gender']=='Female']['Price'].sum())
total_price_by_other='${:,.2f}'.format(purchase_df.loc[purchase_df['Gender']=='Other / Non-Disclosed']['Price'].sum())
```

```python
#Avg Total Purchase per Person
total_purchase_per_male='${:,.2f}'.format(purchase_df.loc[purchase_df['Gender']=='Male']['Price'].sum()/male_players_count)
total_purchase_per_female='${:,.2f}'.format(purchase_df.loc[purchase_df['Gender']=='Female']['Price'].sum()/female_players_count)
total_purchase_per_other='${:,.2f}'.format(purchase_df.loc[purchase_df['Gender']=='Other / Non-Disclosed']['Price'].sum()/other_players_count)
```

```python
#Purchasing Summary
purchasing_summary_df=pd.DataFrame({'Purchase Count':[male_purchase_count,female_purchase_count,other_purchase_count],
                                    'Average Purchase Price':[avg_price_by_male,avg_price_by_female,avg_price_by_other],
                                    'Total Purchase Value':[total_price_by_male,total_price_by_female,total_price_by_other],
                                    'Avg Total Purchase per Person':[total_purchase_per_male,total_purchase_per_female,total_purchase_per_other]},
                                  index=['Male','Female','Other / Non-Disclosed'])
purchasing_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below
* Create a summary data frame to hold the results
* Sort the total purchase value column in descending order
* Optional: give the displayed data cleaner formatting
* Display a preview of the summary data frame

```python
#Group by all purchases by person and get counts
spenders_gb = purchase_df.groupby(['SN'])
counts = spenders_gb.size().to_frame(name='Purchase count')
```

```python
#Aggregate all data, rename columns and display top 5 users
top_spenders=(counts.join(spenders_gb.agg({'Price': 'mean'}).rename(columns={'Price': 'Average Purchase Price'}))
.join(spenders_gb.agg({'Price': 'sum'}).rename(columns={'Price': 'Total Purchase Value'}))
.reset_index()
)
```

```python
#Sort by Purchase count desc and display top 5 results
top_spenders_df=top_spenders.sort_values(by='Purchase count', ascending=False).head(5)
top_spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by all purchases by person and get counts
items_gb = purchase_df.groupby(['Item ID'])
item_counts = items_gb.size().to_frame(name='Purchase count')
```

```python
#Aggregate all data, rename columns and display top 5 items
most_popular_items=(item_counts.join(items_gb.agg({'Item Name':'max'}))
.join(items_gb.agg({'Price': 'max'}).rename(columns={'Price': 'Item Price'}))
.join(items_gb.agg({'Price': 'sum'}).rename(columns={'Price': 'Total Purchase Value'}))
.reset_index()
)
most_popular_items.head()
```

```python
#Sort by Purchase count desc and display top 5 results
most_popular_items_df=most_popular_items.sort_values(by='Purchase count', ascending=False).head(5)
most_popular_items_df
```
